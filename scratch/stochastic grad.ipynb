```python
import numpy as np 
```

```python
class lin_stochastic_grad:
    def __init__(self,lr=0.01,epochs=1000):
        self.lr=lr
        self.epochs=epochs
        self.theta=None

    def fit(self,x,y):
        m,n = x.shape

        x=np.c_[np.ones(m),x]
        self.theta = np.zeros((n+1,1))
        y=y.reshape(-1,1)

        for _ in range(self.epochs):
            # (shuffles order of data)
            indx=np.random.permutation(m)
            x=x[indx]
            y=y[indx]

            # picking random data (x[i:i+!] -> becux we need 2d array for matrix mul)
            for i in range(m):
                xi=x[i:i+1]
                yi=y[i:i+1]

                ypred = xi @ self.theta
                grad = (2/m)*(xi.T @(ypred - yi))
                self.theta-= self.lr*grad
        
    def predict(self,x):
        X = np.c_[np.ones(x.shape[0]),x]
        return X @ self.theta
            
                
            
```

```python
import numpy as np

x = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])

y = np.array([5, 7, 9, 11, 13])   # y = 2x + 3
```

```python
model = lin_stochastic_grad(lr = 0.01 , epochs=5000)
model.fit(x,y)
```

```python
print(model.theta)
```

```python
print(model.predict(np.array([[6]])))
```
