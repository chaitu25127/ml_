```python
import numpy as np
```

```python
class linearRegression:
    def __init__(self,lr=0.01,epochs=1000):
        self.lr= lr
        self.epochs =epochs
        self.theta = None
        self.losshist=[]

    def loss(self,y,ypred):
        return np.mean((ypred-y)**2)

    def fit(self,x,y):
        m,n=x.shape
        #bias (m = no of rows)
        x=np.c_[np.ones(m),x]

        #theta weights(n+1)-> no of features + 1
        self.theta = np.zeros((n+1,1))

        y=y.reshape(-1,1)

        for _ in range(self.epochs):

            ypred = x @ self.theta

            gradient = (2/m) * (x.T @(ypred - y))

            self.theta -= self.lr * gradient 

            loss = self.loss(y,ypred)
            self.losshist.append(loss)
            
    def predict(self,x):
        X = np.c_[np.ones(x.shape[0]),x]
        return X @ self.theta
        
```

```python
import numpy as np

x = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])

y = np.array([5, 7, 9, 11, 13])   # y = 2x + 3
```

```python
model = linearRegression(lr = 0.01 , epochs=5000)
model.fit(x,y)
```

```python
print(model.theta)
```

```python
print(model.predict(np.array([[6]])))
```
