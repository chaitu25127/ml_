```python
import numpy as np
```

```python
class minibatch:
    def __init__(self,lr=0.01,epochs=1000,batchs=32):
        self.lr=lr
        self.epochs = epochs
        self.theta = None
        self.batchs = batchs

    def fit(self,x,y):
        m,n=x.shape

        x=np.c_[np.ones(m),x]

        self.theta = np.ones((n+1 , 1))

        y=y.reshape(-1,1)

        for _ in range(self.epochs):
            indx= np.random.permutation(m)
            x=x[indx]
            y=y[indx]
            
            for start in range (0 , m , self.batchs):

                end = start + self.batchs

                xb=x[start:end]
                yb=y[start:end]

                ypred = xb @ self.theta
                grad = (2/len(xb))*(xb.T @ (ypred - yb))
                self.theta-=self.lr*grad
                
    def predict(self,x):
        X=np.c_[np.ones(x.shape[0]),x]
        return X @ self.theta                
            
```

```python
X = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])

y = np.array([5, 7, 9, 11, 13])   # y = 2x + 3
```

```python
model = minibatch(lr=0.01,epochs=5000,batchs=2)
model.fit(X,y)
```

```python
print(model.theta)
```

```python
print(model.predict(np.array([[6]])))
```
